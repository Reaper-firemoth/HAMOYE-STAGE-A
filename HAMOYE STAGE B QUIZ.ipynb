```python
X_train,X_test,Y_train,Y_test = train_test_split(X_norm,Y_norm, test_size = 0.3, random_state = 42)
```

```python
lR2 = LinearRegression()
lR2.fit(X_train,Y_train)
Y_pred = lR2.predict(X_test)
```

```python
X.columns
```

```python
lR2.coef_
```

```python
def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
linear_model_weights = get_weights_df(lR2, X_train, 'Linear_Model_Weight')
linear_model_weights
```

```python
from sklearn.linear_model import Ridge
Ridg = Ridge(0.4)
L_R = Ridg.fit(X_train,Y_train)
L_R_pred = Ridg.predict(X_test)
```

```python
R_rmse = np.sqrt(mean_squared_error(Y_test,L_R_pred))
R_rmse.round(3)
```

```python
from sklearn.linear_model import Lasso
LS = Lasso(0.001)
L_S = LS.fit(X_train, Y_train)
L_S_pred = LS.predict(X_test)
```

```python
lasso_weights = get_weights_df(L_S, X_train, 'Lasso_Weight')
```
